```python
import numpy as np
from math import sqrt,exp
from matplotlib import pyplot as plt
```

## Local Greens function

```python
def Gii(w):
    delta = 6
    t = 1
    return (0.5 + 2*t/delta) * w / (w**2 - (delta/2 - t)**2) + (0.5 - 2*t/delta) * w / (w**2 - (delta/2 + t)**2)

```

```python
w_range = np.arange(-5,5,0.01)
Gii_all = list(map(Gii, w_range))
plt.plot(w_range, Gii_all)
```

## Spectral function

$$
A(i,\omega) = \left( \frac{1}{2} + \frac{2t}{\Delta} \right)\left[\delta(\omega - \frac{1}{2}\Delta - t) + \delta(\omega + \frac{1}{2}\Delta + t)\right] + \left( \frac{1}{2} - \frac{2t}{\Delta} \right) \left[\delta(\omega - \frac{1}{2}\Delta + t) + \delta(\omega + \frac{1}{2}\Delta - t)\right]
$$

```python
def gauss(x):
    eta = 100
    return np.exp(-(x)**2)/(eta*sqrt(np.pi))

def A_w(w):
    U = 20
    t = 1
    delta = sqrt(U**2 + 16*t**2)
    return (0.5 + 2*t/delta) * (gauss(w - delta/2 - t) + gauss(w + delta/2 + t)) + (0.5 - 2*t/delta) * (gauss(w - delta/2 + t) + gauss(w + delta/2 - t))

w_range = np.arange(-20, 20, 0.001)
A = list(map(A_w, w_range))
plt.plot(w_range, A)
```
